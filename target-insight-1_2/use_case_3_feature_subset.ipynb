```
import mlflow.sklearn
mlflow.sklearn.autolog()


import pandas as pd
df = pd.read_csv('/dbfs/FileStore/tables/hackathon_data.csv')
df = df.fillna(0)
```

```
df.head()
```

```
## Find the mean goal length per goal Assessment type and see the deviation per data point
df['goal_length'] = pd.to_datetime(df['GoalMetDate'])-pd.to_datetime(df['GoalInitiatedDate'])
df['goal_length'] = df['goal_length'].astype(str)
df['goal_length'] = df['goal_length'].str.slice(1, 4)
df['goal_length'] = pd.to_numeric(df['goal_length'], errors='coerce')
df_goal_assessment = df[df['CurrentGoalStatus']=='Met']
df_new = df_goal_assessment.groupby('goalAssessment_encoded', as_index=False)['goal_length'].mean()
df_new = df_new.rename(columns={"goal_length": "mean_goal_len"})
df_new.head(10)
df_stack = pd.concat([df_new, df], axis=1)
df_stack['time_to_goal'] = df_stack['mean_goal_len'] - df_stack['goal_length']
#df_stack[['mean_goal_len','goal_length','time_to_goal']].head(4)
```

```
df_stack['Gender'].head()
```

```
df = df_stack
```

```
## Encode the current goal status
import numpy as np
conditions = [
    (df['CurrentGoalStatus'] == 'Met'),
    (df['CurrentGoalStatus'] == 'On Hold'),
    (df['CurrentGoalStatus'] == 'In Progress'),
    (df['CurrentGoalStatus'] == 'Discontinued')
    ]

# create a list of the values we want to assign for each condition
values = ['1', '2', '3', '0']

# create a new column and use np.select to assign values to it using our lists as arguments
df['current_Goal_Status_encoded'] = np.select(conditions, values)
```

```
#To find the total time taken to reach the goal by multiplying the (Count of new sessions graphed (TrialGroup) per day per year) * (Bins of target periods)
#df['dayBins_graphedByTarget'].unique
df['dayBins_graphedByTarget_new'] = df['dayBins_graphedByTarget'].str.slice(1, 4)
df['dayBins_graphedByTarget_new'] = df['dayBins_graphedByTarget_new'].replace({':':''}, regex=True)
df['dayBins_graphedByTarget_new'].unique
df['dayBins_graphedByTarget_new'] = df['dayBins_graphedByTarget_new'].astype(int)
df['duration_of_trials'] = df['dayBins_graphedByTarget_new']*df['sessionCount_byGoal_byDayYear']
#df['duration_of_trials'].unique()
#df['dayBins_graphedByTarget'] = df['dayBins_graphedByTarget'].astype(int)
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```
categorical_columns = ['Gender']


for col in categorical_columns:
    col_ohe = pd.get_dummies(df[col], prefix=col)
    df = pd.concat((df, col_ohe), axis=1).drop(col, axis=1)
```

```
df['Gender_Female']
```

```
from sklearn.model_selection import train_test_split
df = df.fillna(0)
df['age'] = df['BirthDateYear'].apply(lambda x: 2021-x)



X = df.drop(['binaryGoalMet'], axis=1)
X = df[['duration_of_trials','time_to_goal','goalAssessment_encoded','trialTargetCount_perClient','goalForced_80Percent', 'lag1_percentCorrect','age','Gender_Male','Gender_Female']]
Y = df['binaryGoalMet']
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3)
```

```
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
```

```
predictions = clf.predict(X_test)
```

```
import time
import numpy as np

start_time = time.time()
importances = clf.feature_importances_
std = np.std([
    tree.feature_importances_ for tree in clf.estimators_], axis=0)
elapsed_time = time.time() - start_time

print(f"Elapsed time to compute the importances: "
      f"{elapsed_time:.3f} seconds")
```

```
import pandas as pd
import matplotlib.pyplot as plt
#feature_names = [f'feature {i}' for i in range(X.shape[1])]
col_list = X.columns.values.tolist()
feature_names = col_list
#feature_names = [f'feature {i}' for i in range (X.columns)]
forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```
