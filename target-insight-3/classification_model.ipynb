```
# Import Central Reach Dataset from the CSV uploaded to Azure Databricks Datasets
import pandas as pd
df = pd.read_csv('/dbfs/FileStore/tables/hackathon_data.csv')
```

```
df.head(10)
```

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df.columns
```

```
df.dtypes
```

```
# Pre-process data to create features for the ML models
df['TrialDataYear'] = pd.to_datetime(df['TrialDataDate'])
df['Age'] = df['TrialDataYear'].dt.year - df['BirthDateYear']
df['TrialTargetId_Encoded'] = df['TrialTargetId'].astype('category').cat.codes
df['Gender_Encoded'] = df['Gender'].astype('category').cat.codes
df['GoalDomain_Encoded'] = df['GoalDomain'].astype('category').cat.codes
```

```
# Select the necessaru features for the models
input_df=df[['sessionCount_byGoal_byMonthYear', 'sessionCount_byGoal_byDayYear', 'sessionCount_byGoal_byWeekYear','Gender_Encoded','Age','TrialTargetId_Encoded', 'GoalDomain_Encoded','goalAssessment_encoded','encodedTrialPhase', 'goalForced_80Percent']]
input_df = input_df.replace(np.nan, 0)
input_df.dtypes
```

```
# Input in terms of feature and target variable
X = input_df.drop('goalForced_80Percent', axis=1)
y = input_df['goalForced_80Percent']
```

```
# Create test and train set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```
# Model-1: Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
# Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix, plot_roc_curve
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
# ROC curve
plot_roc_curve(classifier, X_test, y_test)  
pyplot.show()  
```

```
# Feature Importance
importance = classifier.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
```

```
# Plotting feature importance
# pyplot.bar([x for x in range(len(importance))], importance)
# pyplot.show()
feat_importances = pd.Series(importance, index=X_test.columns)
feat_importances.nlargest(20).plot(kind='barh')
```

```
# Model-2: Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)
```

```
# Confusion Matrix
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
# ROC Curve
plot_roc_curve(classifier, X_test, y_test)  
pyplot.show()  
```

```
# Feature Importance
importance = classifier.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
# pyplot.bar([x for x in range(len(importance))], importance)
# pyplot.show()
feat_importances = pd.Series(importance, index=X_test.columns)
feat_importances.nlargest(20).plot(kind='barh')
```

```
# Model-3: Random Forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```
# Confusion Matrix
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
# ROC Curve
plot_roc_curve(classifier, X_test, y_test)  
pyplot.show()  
```

```
# Feature Importance
importance = classifier.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
# pyplot.bar([x for x in range(len(importance))], importance)
# pyplot.show()
feat_importances = pd.Series(importance, index=X_test.columns)
feat_importances.nlargest(20).plot(kind='barh')
```
